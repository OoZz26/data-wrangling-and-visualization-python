## clean

```python
Wrangling_data_clean = Wrangling_data.copy()
```

Problem we need to change the data types

### code

```python
Wrangling_data_clean =Wrangling_data_clean.fillna(-9999999)
integer_cols =['symboling', 'normalized-losses',  'curb-weight', 'engine-size',  'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
Wrangling_data_clean[integer_cols] = Wrangling_data_clean[integer_cols].apply(lambda x :pd.to_numeric(x,errors='coerce').astype(int),axis=0)

float_cols =[  'wheel-base', 'length', 'width', 'height',  'bore', 'stroke']
Wrangling_data_clean[float_cols] = Wrangling_data_clean[float_cols].apply(lambda x :pd.to_numeric(x,errors='coerce').astype(float),axis=0)
```

```python
Wrangling_data_clean.replace(-9999999,np.nan,inplace= True)
```
